# Generate new variables in Dataframe

Selection of single column in the data frame

```python
fb.head(10)
```

```python
fb['Close']
```

```python
fb[['Close','Open']]
```

```python
fb['Price1'] = fb['Close'].shift(-1)
fb
# we move all the columns of close column up by one row
```

```python
fb['PriceDiff'] = fb['Price1']- fb['Close']
fb.head()
# price difference = Close price of tomorrow - Close price of today
```

```python
# Daily return = Price Difference / Close price of today
fb['DailyRet'] = (fb['PriceDiff'] / fb['Close'])
fb.head(10)
```

```python
# Direction 
def trend(num):
    if (num > 0): 
        return 1
    else:
        return -1

fb['Direction'] = fb['PriceDiff'].apply(lambda x:trend(x))
fb.head(10)
```

# Use of list comprehension

```python
# other method 

fb['example_Direction'] = [1 if fb.loc[date, 'PriceDiff']>0 else -1 for date in fb.index]
fb.head(10)

# shift(-1) will shift the entire column 1 step up ignoring the first value
# shift(1) will shift the entire column down leaving the top column vacant
# the number inside the shift will increment or decrement the number of rows in the given dataframe
```

```python
# moving average is a widely used signal in stock trading which is loading average price over a defined number of days
```

```python
fb['movingAverage'] = (fb['Close']+fb['Close'].shift(1)+fb['Close'].shift(2))/3
# average of close price today, yesterday and the day before yesterday
fb.head(10)
```

```python
fb['MA40'] = fb['Close'].rolling(40).mean()
fb['MA200'] =fb['Close'].rolling(200).mean()
fb
```

```python
fb['Close'].plot() # coloured in blue
fb['MA40'].plot() # coloured in orange
fb['MA200'].plot() # coloured in green
```

Observations.
1 Moving average 40 is more closely related to the trend.
2 Moving average 200 is less closely related to the trend.

Hence we call the moving average 40 as the fast signal and the moving average 200 as the slow signal

if MA40 > MA200, Some traders who are called trend following traders believe the stock prices will move upwards for a while

# Simple Trading Strategy

```python
ms = pd.read_csv('microsoft.csv')
ms['MA10'] = ms['Close'].rolling(10).mean() # fast signal 
ms['MA50'] = ms['Close'].rolling(50).mean() # slow signal

```

```python
ms['Close'].plot(legend = True)
ms['MA10'].plot(legend = True)
ms['MA50'].plot(legend = True)
```

if MA10 > MA50 then it is believed the price of the stock will grow in the next few days otherwise the price will go down

```python
ms['Shares'] = [1 if ms.loc[date,'MA10']>ms.loc[date,'MA50'] else 0 for date in ms.index]
ms
```

```python
# daily profit
ms['Close1'] = ms['Close'].shift(-1)
ms['Profit'] = [ms.loc[date,'Close1']-ms.loc[date,'Close'] if ms.loc[date,'Shares']==1 else 0 for date in ms.index]
ms.head(10)
```

```python
ms['Profit'].plot()
```

```python
# cummulative wealth 
ms['wealth'] =ms['Profit'].cumsum()
ms['wealth'].max()
ms.tail(10)
```

```python
print(' Total money you win is ',ms.loc[ms.index[-2],'wealth'])
print(' Total money you lose is ',ms.loc[ms.index[0],'Close'])
```
